# Data visualization using scatter and histogram

Histogram-

Histogram is one of the best solution which lets us know the total Number of car manufactured.

```python
data.Model.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Model');
```

Scatter plot-

To find the correlation between two variables.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(data['HP'], data['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

 Heatmap-

```python
plt.figure(figsize=(10,5))
c = data.corr()
sns.heatmap(c, cmap="BrBG", annot=True )
```
